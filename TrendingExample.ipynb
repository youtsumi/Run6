```python
from astropy.io import ascii
from astropy.time import Time, TimezoneInfo
import pylab
import astropy.units as u
```

```python
output=!~/src/mutils/bin/trender.py \
    --start "2023/10/13 14:30 PDT" \
    --text --site slac \
    --timebin \
    --dur 2d -- 'vacuum/Cryo/CryoVac'
output
```

```python
table=ascii.read(output)

x=Time(table["col1"]/1000,format="unix")
y=table["col2"]
```

```python
pdt = TimezoneInfo(utc_offset=-7*u.hour)
pylab.plot(x.to_datetime(timezone=pdt),y,"o",label="data")
pylab.semilogy()
```
